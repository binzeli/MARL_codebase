```python
import matplotlib.pyplot as plt
import os

# Your trajectory data
trajectories = [[(0, 0), (0, 0)], [(0, 1), (1, 0)], [(0, 2), (2, 0)], [(0, 2), (2, 1)], [(0, 2), (2, 2)], [(0, 3), (3, 2)], [(0, 3), (3, 3)], [(0, 4), (4, 3)]]

# Create a directory for frames
os.makedirs('frames', exist_ok=True)

# Generate each frame
for i in range(len(trajectories)):
    fig, ax = plt.subplots()
    ax.plot([p[0][0] for p in trajectories[:i+1]], [p[0][1] for p in trajectories[:i+1]], 'bo-')  # Agent 1
    ax.plot([p[1][0] for p in trajectories[:i+1]], [p[1][1] for p in trajectories[:i+1]], 'ro-')  # Agent 2
    ax.set_xlim(-1, 5)
    ax.set_ylim(-1, 5)
    ax.set_aspect('equal')

    # Save frame
    plt.savefig(f'frames/frame_{i:04d}.png')
    plt.close()

```
